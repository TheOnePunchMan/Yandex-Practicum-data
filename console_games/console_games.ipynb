## Описание проекта

Работаем в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). **Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.  
Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными.**  
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

## Предобработка данных

### Приведение названий столбцов

Приведем к нижнему регистру названия столбцов:

````python
df.columns = map(str.lower, df.columns)
df = df.rename(columns=
        {
        'na_sales':'na_sales_mil',
        'eu_sales':'eu_sales_mil',
        'jp_sales':'jp_sales_mil',
        'other_sales':'other_sales_mil',
        }
              )
df.head()
````

### Обработка пропусков

Выведем структуру пропусков:

````python
def nan_struct(df=df):
    nan_count_df = df.isna().agg(['sum', 'mean'])
    nan_count_df.loc['mean',:] = (nan_count_df.loc['mean',:] * 100).round(2).astype(str) + '%'
    nan_count_df.index = ['nan_counts', 'nan_part']
    return nan_count_df.T [nan_count_df.loc['nan_counts',:] > 0]

nan_struct()
````

Есть предположение, что пропуски в названии игры и её жанре связаны, проверим:

````python
df[ df['name'].isna() ]
````

В этих двух строчках почти все значения пропущены, вклад в данные они не делают, поэтому удалим их:

````python
df = df[ ~df['name'].isna() ].reset_index(drop=True)
df.tail()
````

````python
nan_struct(df)
````

Посмотрим на зависимость в пропусках между столбцами `user_score`, `critic_score` и `rating`:

````python
df[ df['user_score'].isna() & df['critic_score'].isna() & df['rating'].isna()]
````

Видно, что общих пропусков 6619, то есть по большей части значения пропущены в трех столбцах сразу.  
Скорее всего игроки, как и критики не успели еще сделать обзор этих игр, так как они либо относительно недавно вышли, либо малопопулярны, либо уже не нуждаются в оценке.

Удалять такой объем данных не очень разумно, поэтому заменим все пропущенные значения в столбце `rating` значениями 'unknown':

````python
df['rating'] = df['rating'].fillna('unknown')
````

А также заменим все пропущенные пары: оценка пользователя - оценка эксперта техническими значениями -10:

````python
df.loc[ df['user_score'].isna() & df['critic_score'].isna(),
       ['user_score', 'critic_score']] = df.loc[
    df['user_score'].isna() & df['critic_score'].isna(),
    ['user_score', 'critic_score']].fillna(-10)
````

````python
nan_struct(df)
````

Заполним оставшиеся оценки пользователей на основании оценок критиков:

````python
df.loc[ df['user_score'].isna(), 'user_score'] = (
    df.loc[ df['user_score'].isna(), 'critic_score'] / 10 ).round()
````

````python
nan_struct(df)
````

Посмотрим, есть ли в столбце `user_score` зависимость между оценками критиков и оценкой пользователей "tbd".

````python
df[ df['user_score'] == 'tbd' ]['critic_score'].hist(bins=20);
````

Как и ожидалось, пользователи поставили оценку "tbd" (to be determined), так как еще не успели оценить игру.  
Так как данные столбца ранговые, заполним техническим значением, например -10, все ячейки "tbd":

````python
df['user_score'] = df['user_score'].where(df['user_score'] != "tbd", -10)
df['user_score'] = df['user_score'].astype('float')
df.head()
````

Теперь заполним оставшиеся оценки критиков на основании оценок пользователей (при этом не домножаем значения -10):

````python
df['critic_score'] = df['critic_score'].fillna(
    df['user_score'].transform(lambda x: x * 10 if x > 0 else x))
````

````python
nan_struct(df)
````

````python
df[ df['year_of_release'].isna() ]
````

Пропуски в столбце `year_of_release` частично можно заполнить с помощью столбца `name`:

Напишем функцию, которая выделяет год из последнего элемента каждой ячейки столбца `name`

````python
def select_last_number(el):
    try:
        if int(el[-1]) in range(1980, 2022):
            return int(el[-1])
    except ValueError:
        pass
    
    
sel = df['name'].str.split().apply(select_last_number).dropna()
sel = sel.astype('int')
sel
````

````python
df['year_of_release'] = df['year_of_release'].fillna(sel)
nan_struct(df)
````

Пропусков стало меньше. Проверим значимость строчек для анализа с пропусками в годе выпуска:

````python
df[ df['year_of_release'].isna() ].query(
    'na_sales_mil + eu_sales_mil + jp_sales_mil + other_sales_mil > 1').head(10)
````

Присутствует достаточное количество игр с более, чем миллионной аудиторией $\Rightarrow$ строки удалять нельзя.  
Тогда заменим значение года медианным значением года платформы, на которой работает игра.

````python
df['year_of_release'] = df['year_of_release'].fillna(
    df.groupby('platform')['year_of_release'].transform('median')
)
````

````python
df.isna().sum().to_frame().T
````

## Проверка гипотез

**Гипотеза 1:** Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.  
За нулевую гипотезу примем равенство средних рейтингов данных платформ. Так как выборки независимы:

````python
alpha = 0.05

results = st.ttest_ind(df_actual[(df_actual['platform']=='PC') & (df_actual['user_score']>0)]['user_score'],
                       df_actual[(df_actual['platform']=='XOne') & (df_actual['user_score']>0)]['user_score'],
                       equal_var=False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
````

p-value принимает довольно большое значение $\Rightarrow$ не можем отвергнуть нулевую гипотезу.  
Считаем, что средние пользовательские рейтинги платформ Xbox One и PC равны между собой.

**Гипотеза 2:** Средние пользовательские рейтинги жанров Action и Sports разные.  
За нулевую гипотезу примем факт равенства средних пользовательских рейтингов на для этих жанров.  
Тогда альтернативной гипотезой будет отсутствие равенства средних значений рейтингов.  
Тогда для двух генеральных совокупностей (в предположении, что выборки независимы) воспользуемся критерием Стьюдента (так как нулевая гипотеза на равенство средних значений, а распределения напоминают нормальные со скошенностью влево):

````python
fig = plt.figure(figsize=(14,6))
fig.subplots_adjust(hspace=0.4, wspace=0.4)

plt.subplot(1,2,1)
df_actual[df_actual['genre']=='Action']['user_score'].plot(kind='hist', bins=100)
plt.xlim(0,10)
plt.ylim(0,70)
plt.ylabel('Частота')
plt.title('Action user_score')

plt.subplot(1,2,2)
df_actual[df_actual['genre']=='Sports']['user_score'].plot(kind='hist', bins=100)
plt.xlim(0,10)
plt.ylim(0,20)
plt.ylabel('Частота')
plt.title('Sports user_score');
````

````python
alpha = 0.05

results = st.ttest_ind(df_actual[(df_actual['genre']=='Action') & (df_actual['user_score']>0)]['user_score'],
                       df_actual[(df_actual['genre']=='Sports') & (df_actual['user_score']>0)]['user_score'],
                       equal_var=False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
````

Так как p-value $\approx$ 0 $\Rightarrow$ отвергнем нулевую гипотезу $\Rightarrow$ средние пользовательские рейтинги жанров Action и Sports разные.
