## Предобработка данных

### Обработаем пропуски

```python
df.isna().sum()[ df.isna().sum() > 0]
```

✔ Выведем структуру пропусков:

```python
def nan_struct(df=df):
    nan_count_df = df.isna().agg(['sum', 'mean'])
    nan_count_df.loc['mean',:] = (nan_count_df.loc['mean',:] * 100).round(2).astype(str) + '%'
    nan_count_df.index = ['nan_counts', 'nan_part']
    return nan_count_df.T [nan_count_df.loc['nan_counts',:] > 0]

nan_struct()
```

Проверим, есть ли зависимость между пустыми значениями в столбцах с количеством парков и расстоянием до ближайшего парка:

```python
df_parks = df.copy()
for el in ['parks_around3000', 'parks_nearest']:
    df_parks = df_parks[ df_parks[el].isna() ]
len(df_parks)
```

```python
df_parks.head()
```

Аналогично для водоемов:

```python
df_ponds = df.copy()
for el in ['ponds_around3000', 'ponds_nearest']:
    df_ponds = df_ponds[ df_ponds[el].isna() ]
len(df_ponds)
```

```python
df_ponds.sample(7)
```

Видно, что столбцы попарно не заполняются, когда в `parks_around3000` или `parks_nearest` пропущены значения.  
Аналогично для `ponds_around3000` или `ponds_nearest`. Значит, скорее всего в окрестности недвижимости нет парков/прудов соответственно, поэтому смело заменим столбцы попарно нулями.  
Также заменим количество балконов нулями, так как если они не указаны, то их скорее всего нет.

```python
df['balcony'].fillna(0, inplace=True)
df_ponds_mask = (df['ponds_around3000'].isna()) & (df['ponds_nearest'].isna())
df.loc[df_ponds_mask,
       ['ponds_around3000', 'ponds_nearest']] = df.loc[df_ponds_mask,
                                                       ['ponds_around3000', 'ponds_nearest']].fillna(value=0)
#можно ли тут как-то сократить конструкцию?
df_parks_mask = (df['parks_around3000'].isna()) & (df['parks_nearest'].isna())
df.loc[df_parks_mask,
       ['parks_around3000', 'parks_nearest']] = df.loc[df_parks_mask,
                                                       ['parks_around3000', 'parks_nearest']].fillna(value=0)
df.isna().sum()
```

```python
df.shape
```

В категориальных переменных заменим пропущенные значения на 'unknown'

```python
df['is_apartment'] = df['is_apartment'].fillna('unknown')
df.sample(7)
```

### Проверим данные на дубликаты

```python
df['locality_name'].value_counts()
```

```python
df['locality_name'] = df['locality_name'].str.lower()
df.duplicated().sum()
```

Обзор столбца даёт понимание, что присутствуют населенные пункты, записанные с разным типом - "деревня\село\поселок". Можно избавиться от всего, что идет до названия, а критерием начала названия считать заглавную букву (однако таких городов немного).

Явных дубликатов нет.

Посчитаем дубликаты после лемматизации буквенного столбца `locality_name`

```python
df['locality_name'] = df['locality_name'].astype('str')
from pymystem3 import Mystem
m = Mystem()
def lemmatize_text(row):
    loc_name = row['locality_name']
    return ' '.join(m.lemmatize(loc_name))
df['locality_name'] = df.apply(lemmatize_text, axis=1)

df.duplicated().sum()
```

Дубликатов не обнаружено.

### Приведем к нужным типам данных

Столбец `is_apartment` на логический тип данных менять не будем, так как теперь столбец содержит большое число значений 'unknown', которые могут превратиться при переводе в *True*.<br>   
Заменим многие другие столбцы, не содержащие NaN, на целые числа(где это полагается по смыслу). 

```python
for el in ['last_price', 'airports_nearest', 'cityCenters_nearest']:
    if len(df[df[el].isna()]) == 0:
        df[el] = df[el].astype('uint32') #4 байта на число до 4 миллиардов хватит
```

```python
for el in ['parks_nearest', 'ponds_nearest', 'days_exposition']:
    if len(df[df[el].isna()]) == 0:
        df[el] = df[el].astype('uint16') #2 байта хватит до 65000
```

```python
for el in ['floors_total', 'balcony', 'parks_around3000', 
           'ponds_around3000', 'total_images', 'rooms', 'floor']:
    if len(df[df[el].isna()]) == 0:
        df[el] = df[el].astype('uint8') #1 до 255
```

```python
for el in ['ceiling_height', 'living_area', 'kitchen_area']:
    pd.to_numeric(df[el], errors='ignore', downcast='float')
```

```python
df.info()
```

Приведем столбец `first_day_exposition` к типу datetime, чтобы впоследствии проще работать с округлением и визуализацией.

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
df.dtypes
```

## Расчёты и добавление результатов в таблицу

Цена квадратного метра квартиры:

```python
df['price_per_square'] = df['last_price'] / df['total_area'] 
```

День недели, месяц, год публикации:

```python
df['weekday'] = df['first_day_exposition'].dt.weekday
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year
```

Этаж квартиры (в виде категориальной переменной):

```python
def floor_categorisation(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    return 'другой'
    
df['floor_cat'] = df.apply(floor_categorisation, axis=1)
```

Соотношение жилой площади к общей, а также кухонной площади к общей:

```python
df['living_to_total'] = df['living_area'] / df['total_area']
df['kitchen_to_total'] = df['kitchen_area'] / df['total_area']
df.head()
```

### Избавление от особенностей в данных

Выбросим все объекты из пунктов **4.1**, **4.2**, где значения столбцов выходят за 1.5 межквартильного размаха относительно верхней и нижней квартилей:

```python
df = df.loc[df['total_area'] < 190, :]
df = df.loc[df['living_area'] < 90, :]
df = df.loc[df['kitchen_area'] < 20, :]
df = df.loc[(df['living_to_total'] < 0.9) & (df['living_to_total'] > 0.2), :]
df = df.loc[df['kitchen_to_total'] < 0.4, :]
df = df.loc[(df['price_per_square'] < 175000) & (df['price_per_square'] > 20000), :]
df = df.loc[df['last_price'] < 1.25e7, :]
df = df.loc[(df['ceiling_height'] < 3.5) & (df['ceiling_height'] > 2), :]
df = df.loc[df['rooms'] < 7, :]
df = df.query('10 < days_exposition < 514')

df.sample(10, random_state=42)
```

#### Анализ зависимости цены от близости к центру

Так как после 3км от центра цена резко меняется - будем считать центром область радиусом 3км.

#### Выделение центральной области

Выделим сегмент квартир в центре. Проанализируем эту территорию и изучим следующие параметры: площадь, цена, число комнат, высота потолков.

```python
st_petersburg_center = st_petersburg_df.query('distance_km <= 3')
st_petersburg_center.sample(8)
```

```python
nrows, ncols = 3, 3
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(14,16))

plot_list = ['total_area', 'living_area', 'kitchen_area', 'living_to_total',
             'kitchen_to_total', 'price_per_square', 'last_price', 'ceiling_height', 'rooms']
i = 0
for fig_i in range(nrows):
    for fig_j in range(ncols):
        axes[fig_i,fig_j].set_xlabel(plot_list[i])
        st_petersburg_center[ plot_list[i] ].plot(kind='hist', bins=25, ax=axes[fig_i,fig_j])
        i += 1
```

Распределения величин теперь напоминают нормальное распределение.  
Гистограмма стоимости имеет небольшой пик в области больших значений.

Выделим факторы, которые влияют на стоимость квартиры (число комнат, этаж, удалённость от центра, дата размещения объявления). Сделаем выводы. Отличаются ли они от общих выводов по всему городу?

```python
param_spb_center_df = st_petersburg_center[['last_price', 'total_area', 'rooms', 'floor', 'cityCenters_nearest']]
param_spb_center_time_df = st_petersburg_center[['last_price', 'weekday', 'month', 'year']]

nrows, ncols = 2, 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(16,12))

i = 0
for fig_i in range(nrows):
    for fig_j in range(ncols):
        param_spb_center_df.plot(x=param_list[i], y='last_price', kind='hexbin', gridsize=50, sharex=False, ax=axes[fig_i,fig_j])
        i += 1
```

В среднем видна положительная корреляция между числом комнат и ценой, а также между ценой и площадью.  
В центральной части города не наблюдается явной зависимости цены от близости к центру, наоборот прослеживается небольшая отрицательная корреляция между этими величинами.

Рассмотрим, как зависит цена от этажа:

```python
price_floor_cat_center_df = st_petersburg_center.pivot_table(
    index='floor_cat', values='last_price', aggfunc=['median', 'count']
)
price_floor_cat_center_df.columns = ['median_price', 'num_floors']
price_floor_cat_center_df.plot(y='median_price', kind='bar', color='c');
```

Квартиры на первом этаже в центре Санкт-Петербурга стоят меньше, что на других, а на последнем этаже цена самая высокая.

Изучим зависимость цены от даты выложенного объявления:

```python
param_spb_center_time_df.plot(x='weekday', y='last_price', kind='hexbin', sharex=False, gridsize=12);
```

```python
param_spb_center_time_df.plot(x='month', y='last_price', kind='hexbin', sharex=False, gridsize=10);
```

Цена не зависит от даты публикации, видно только, что в 2017, 2018 году было опубликована основная часть всех объявлений по ценам в диапазоне от 3.5 млн до 12 млн.

В центре Санкт-Петербурга цена квартиры прямо пропорциональна её площади, числу комнат, при этом имеется небольшая обратная связь между ценой и расстоянием до центра.
